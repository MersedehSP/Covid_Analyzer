```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv("Resources/all-states-history (1).csv")
raw_data
```

```python
#extract columns we will need
covid_df = raw_data.loc[:,['date', 'state', 'death', 'hospitalized', 'totalTestResults']]
covid_df = covid_df.rename(columns={"date":"Date", "state":"State", "death":"Number of Deaths", 
                                   "hospitalized":"Number of Hospitalization", "totalTestResults": "Total Tests Taken"})
covid_df
```

```python
# Ask the user how many numbers to loop through
national_df = covid_df.loc[:,['State', 'Number of Deaths','Number of Hospitalization', 'Total Tests Taken']]
ask_state_vs_national = str(input("Do you want to see National Data (y/n)"))


if ask_state_vs_national == str("y"):
    print("***Here is the summary of National Data ***")
    group_df = national_df.groupby(["State"])
    display_df =group_df.sum()
    display_df = display_df.astype(int)
    print(display_df)
    more_data = str(input("***Do you want to see more data (y/n)***"))
    if more_data == str("y"):
        display_df["Total Infected"] = display_df["Number of Deaths"] + display_df["Number of Hospitalization"]
        display_df["Infection Percentage"] = (display_df["Total Infected"]/display_df["Total Tests Taken"])*100
        display_df["Infection Percentage"] = display_df["Infection Percentage"].map("{:.2f}%".format)
        display_df["Total Cured"] = display_df["Number of Hospitalization"] - display_df["Number of Deaths"]
        print(display_df)
else:
    print("Ok you want to see State Wide Data")
    state_name = str(input("***Please enter the name of the state in abbreviation example: NewJersey is NJ****"))
    unique = national_df["State"].unique()
    print(unique)
    while state_name == str(unique.all()):
        print("OK")
    
```
